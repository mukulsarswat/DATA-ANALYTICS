```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6,7]
y=[15,23,21,54,12,32,43]
y1=[25,28,30,25,34,24,20]
y2=[15,18,10,15,14,14,20]

plt.figure(figsize=(10,6))

plt.subplot(2,4,1)
plt.title("Male v/s Age")
plt.plot(x,y)

plt.subplot(2,4,2)
plt.title("Female v/s Age")
plt.bar(x,y1)

plt.subplot(2,4,3)
plt.title("Children")
plt.scatter(x,y2)

plt.suptitle("Villagers")

plt.tight_layout()
plt.savefig("Villagers_Chart.png")  #  Save before show
plt.show()
```

```python
# Functions in Numpy for Addition,Subtract,Multiply,Square,exp,Divide
a=[111,1111,11111]
b=[1111,11111,11111]
arr1=np.array([a,b])
arr2=np.array([b,a])
print(np.add(arr1,arr2))# print(arr1+arr2)

print(np.subtract(arr1,arr2))  #print(arr1-arr2)
print(np.square(arr1))
print(np.power(arr1,3))
# print(arr2)
```

```python
# Combining and Splitting The Array --> Concatenate Fun to combine
print(np.concatenate([arr1,arr2],axis=1))
print(np.hstack([arr1,arr2]))
print(np.vstack([arr1,arr2]))
arrr=[1,2,3,4,5,6]
aa=np.array(arrr)
print(aa)
b=np.array_split(aa,2)
print(b[0])
```

```python
#Adding and removing elements in the Array
a=np.array([20,40,60,80])
print(a)
a=np.append(a,[90,100])
print(a)
a=np.insert(a,[0,2],55) # We can add val at multiple locations
print(a)
```

```python
# SORT,FILTER and Search
w=np.array([3,4,5,2,1,7,6,54])
print(w)
ww=np.sort(w)
print(np.where(w%3==0))
print(np.searchsorted(ww,4))
fa=[True,False,True,True,True,True,True,True] # Filtering
fb=w>3
fc=w%2==0
print(w[fb])
print(w[fa])
print(w[fc])
```
